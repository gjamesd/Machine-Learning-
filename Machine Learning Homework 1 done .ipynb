## Trying to do something cool with both sql and csv 


```python
import sqlite3 as sql
import csv
#This function creates a db file from a csv so we can manipulate the data in sql
def earthquakes_db(db_file="earthquakes.db", 
                   data_file="us_earthquakes.csv"):
    """
    Parameters:
        db_file (str): The name of the database file.
        data_file (str): The name of a csv file containing data for the
            USEarthquakes table.
    """
    #opening file within database 
    print(data_file)
    with open(data_file, 'r') as infile:
        rows = list(csv.reader(infile))
        
    #running basi sql 

    with sql.connect(db_file) as conn:
        cur = conn.cursor()

        # Recreate the tables.
        cur.execute("DROP TABLE IF EXISTS USEarthquakes;")
        cur.execute("CREATE TABLE USEarthquakes (Year INTEGER, Month INTEGER, Day INTEGER, Hour INTEGER, Minute INTEGER, Second INTEGER, Latitude REAL, Longitude REAL, Magnitude REAL);")

        # Populate the table.
        cur.executemany("INSERT INTO USEarthquakes VALUES (?,?,?,?,?,?,?,?,?);", rows)

        cur.execute("DELETE FROM USEarthquakes WHERE Magnitude == 0")
        cur.execute("UPDATE USEarthquakes SET Day='NULL', Hour='NULL', Minute='NULL', Second='NULL' WHERE Day == 0")
        
        for row in cur.execute("SELECT * FROM USEarthquakes"):
            print(row)
        conn.commit()
conn.close()
if __name__=="__main__":
    earthquakes_db(db_file="earthquakes.db", data_file="us_earthquakes.csv")
```
